```python
#10^-5
import pandas as pd
import random
import matplotlib
import numpy
import matplotlib.pyplot as plt

mark = random.Random()

efforts = 150 # number of 
actions = 30 # number of
statistics = []
for i in range(efforts):
    marks = []
    for j in range(actions):
        marks.append(mark.randint(4, 5))
    statistics.append(sum(marks)/actions)
plt.plot(list(range(efforts)), statistics)
plt.title("Student marks")
plt.show
```

```python
stat = pd.DataFrame(data={
    'statistics': statistics
})

stat.to_csv("statistics.csv")
```
